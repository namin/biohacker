```python
import random

# Define a simple synthesizer
class ProgramSynthesizer:
    def __init__(self):
        self.program_templates = {
            "add": "def add(x, y): return x + y",
            "subtract": "def subtract(x, y): return x - y",
        }
        self.template_weights = {
            "add": 1.0,
            "subtract": 1.0,
        }
        self.feedback_history = []

    def synthesize_program(self, problem):
        selected_template = self.select_template()
        return self.program_templates[selected_template]

    def select_template(self):
        templates = list(self.template_weights.keys())
        weights = list(self.template_weights.values())
        return random.choices(templates, weights)[0]

    def receive_feedback(self, problem, feedback):
        self.feedback_history.append((problem, feedback))
        self.update_template_weights(feedback)

    def update_template_weights(self, feedback):
        if feedback == "positive":
            for template in self.template_weights:
                self.template_weights[template] *= 1.1
        elif feedback == "negative":
            for template in self.template_weights:
                self.template_weights[template] *= 0.9
        self.normalize_weights()

    def normalize_weights(self):
        total_weight = sum(self.template_weights.values())
        for template in self.template_weights:
            self.template_weights[template] /= total_weight

# Example usage
synthesizer = ProgramSynthesizer()
problem = "Write a program that adds two numbers."

# Synthesize a program
generated_program = synthesizer.synthesize_program(problem)
print("Generated program:")
print(generated_program)

# Simulate user feedback
user_feedback = "positive"
synthesizer.receive_feedback(problem, user_feedback)

# Synthesize another program
generated_program = synthesizer.synthesize_program(problem)
print("\nGenerated program after receiving feedback:")
print(generated_program)
```

This simple example synthesizes a program for a given problem by selecting a template based on the current weights.
